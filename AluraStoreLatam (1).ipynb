### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda1["Precio"] = pd.to_numeric(tienda1["Precio"], errors="coerce")
tienda2["Precio"] = pd.to_numeric(tienda2["Precio"], errors="coerce")
tienda3["Precio"] = pd.to_numeric(tienda3["Precio"], errors="coerce")
tienda4["Precio"] = pd.to_numeric(tienda4["Precio"], errors="coerce")

ingresos_tienda1 = tienda1["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()

print("Ingresos Tienda 1:", ingresos_tienda1)
print("Ingresos Tienda 2:", ingresos_tienda2)
print("Ingresos Tienda 3:", ingresos_tienda3)
print("Ingresos Tienda 4:", ingresos_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = tienda1["Categoría del Producto"].value_counts()
ventas_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

print("Ventas por categoría - Tienda 1")
print(ventas_categoria_tienda1)
print("\n")

print("Ventas por categoría - Tienda 2")
print(ventas_categoria_tienda2)
print("\n")

print("Ventas por categoría - Tienda 3")
print(ventas_categoria_tienda3)
print("\n")

print("Ventas por categoría - Tienda 4")
print(ventas_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
tienda1["Calificación"] = pd.to_numeric(tienda1["Calificación"], errors="coerce")
tienda2["Calificación"] = pd.to_numeric(tienda2["Calificación"], errors="coerce")
tienda3["Calificación"] = pd.to_numeric(tienda3["Calificación"], errors="coerce")
tienda4["Calificación"] = pd.to_numeric(tienda4["Calificación"], errors="coerce")

calificacion_tienda1 = tienda1["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

print("Calificación promedio - Tienda 1:", round(calificacion_tienda1, 2))
print("Calificación promedio - Tienda 2:", round(calificacion_tienda2, 2))
print("Calificación promedio - Tienda 3:", round(calificacion_tienda3, 2))
print("Calificación promedio - Tienda 4:", round(calificacion_tienda4, 2))
```

# 4. Productos más y menos vendidos

```python
mas_vendido_tienda1 = tienda1["Producto"].value_counts().idxmax()
mas_vendido_tienda2 = tienda2["Producto"].value_counts().idxmax()
mas_vendido_tienda3 = tienda3["Producto"].value_counts().idxmax()
mas_vendido_tienda4 = tienda4["Producto"].value_counts().idxmax()

menos_vendido_tienda1 = tienda1["Producto"].value_counts().idxmin()
menos_vendido_tienda2 = tienda2["Producto"].value_counts().idxmin()
menos_vendido_tienda3 = tienda3["Producto"].value_counts().idxmin()
menos_vendido_tienda4 = tienda4["Producto"].value_counts().idxmin()

print("Tienda 1 - Más vendido:", mas_vendido_tienda1)
print("Tienda 1 - Menos vendido:", menos_vendido_tienda1)
print()

print("Tienda 2 - Más vendido:", mas_vendido_tienda2)
print("Tienda 2 - Menos vendido:", menos_vendido_tienda2)
print()

print("Tienda 3 - Más vendido:", mas_vendido_tienda3)
print("Tienda 3 - Menos vendido:", menos_vendido_tienda3)
print()

print("Tienda 4 - Más vendido:", mas_vendido_tienda4)
print("Tienda 4 - Menos vendido:", menos_vendido_tienda4)
```

```python
import matplotlib.pyplot as plt

def graficar_top5(df, nombre_tienda):
    top5 = df["Producto"].value_counts().head(5)

    plt.figure()
    plt.bar(top5.index, top5.values)
    plt.title(f"Top 5 productos más vendidos - {nombre_tienda}")
    plt.xticks(rotation=45)
    plt.ylabel("Cantidad de ventas")
    plt.show()

graficar_top5(tienda1, "Tienda 1")
graficar_top5(tienda2, "Tienda 2")
graficar_top5(tienda3, "Tienda 3")
graficar_top5(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
tienda1["Costo de envío"] = pd.to_numeric(tienda1["Costo de envío"], errors="coerce")
tienda2["Costo de envío"] = pd.to_numeric(tienda2["Costo de envío"], errors="coerce")
tienda3["Costo de envío"] = pd.to_numeric(tienda3["Costo de envío"], errors="coerce")
tienda4["Costo de envío"] = pd.to_numeric(tienda4["Costo de envío"], errors="coerce")

envio_prom_tienda1 = tienda1["Costo de envío"].mean()
envio_prom_tienda2 = tienda2["Costo de envío"].mean()
envio_prom_tienda3 = tienda3["Costo de envío"].mean()
envio_prom_tienda4 = tienda4["Costo de envío"].mean()

print("Costo de envío promedio - Tienda 1:", round(envio_prom_tienda1, 2))
print("Costo de envío promedio - Tienda 2:", round(envio_prom_tienda2, 2))
print("Costo de envío promedio - Tienda 3:", round(envio_prom_tienda3, 2))
print("Costo de envío promedio - Tienda 4:", round(envio_prom_tienda4, 2))
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envios_promedio = [
    envio_prom_tienda1,
    envio_prom_tienda2,
    envio_prom_tienda3,
    envio_prom_tienda4
]

plt.figure()
plt.bar(tiendas, envios_promedio)
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo de envío promedio")
plt.xticks(rotation=45)
plt.show()
```

```python
# Informe Final - Análisis de Desempeño de Tiendas Alura Store

El presente análisis tiene como objetivo evaluar el desempeño de las cuatro tiendas de la cadena Alura Store, con el fin de recomendar cuál de ellas debería vender el Sr. Juan para iniciar un nuevo emprendimiento.

Para ello, se analizaron métricas clave como:

- Ingresos totales.
- Categorías de productos más y menos vendidas.
- Calificaciones promedio de los clientes.
- Productos más y menos vendidos.
- Costo de envío promedio.

A partir de estos indicadores y de las visualizaciones generadas, se realizó una comparación integral entre las tiendas.

### Ingresos Totales

El análisis de los ingresos mostró que la Tienda 4 presenta el menor ingreso total en comparación con las demás tiendas. Esto indica un menor volumen de facturación y, por tanto, un desempeño comercial inferior.

Las Tiendas 1, 2 y 3 presentan ingresos superiores, destacándose especialmente la Tienda 1 con la mayor facturación total.

### Categorías Más Vendidas

En todas las tiendas, la categoría con mayor generación de ingresos es Electrónicos, lo que demuestra que esta línea de productos es estratégica para la cadena.

Sin embargo, aunque la Tienda 4 también vende principalmente Electrónicos, su volumen total de ventas es inferior al de las demás tiendas.

### Calificaciones Promedio

Las calificaciones promedio muestran que:

- La Tienda 3 posee la mejor evaluación de clientes.
- La Tienda 2 también mantiene un nivel alto de satisfacción.
- La Tienda 4 presenta una calificación aceptable, pero no superior a las otras tiendas.
- La Tienda 1 tiene la calificación más baja.

Aunque la Tienda 4 no tiene la peor calificación, su desempeño en ingresos no compensa esta ventaja moderada.

### Productos Más y Menos Vendidos

Cada tienda presenta productos destacados, especialmente en la categoría de Electrónicos. Sin embargo, la Tienda 4 no muestra un producto líder que genere un impacto significativo en comparación con los productos estrella de otras tiendas.

### Costo de Envío Promedio

La Tienda 4 tiene el menor costo de envío promedio, lo cual podría considerarse una ventaja competitiva.

No obstante, esta reducción en el costo de envío no se traduce en mayores ingresos ni en un liderazgo en ventas, lo que indica que el problema principal no está en el envío sino en el rendimiento comercial general.

## 3. Conclusión y Recomendación

Con base en el análisis integral de ingresos, volumen de ventas, satisfacción del cliente, productos destacados y costos de envío, se recomienda que el Sr. Juan venda la Tienda 4.

Justificación:

- Es la tienda con menor ingreso total.
- No lidera en calificación promedio.
- No presenta un producto dominante que impulse significativamente sus ventas.
- A pesar de tener el menor costo de envío, esto no se refleja en un mejor desempeño financiero.

Las Tiendas 1, 2 y 3 muestran un rendimiento superior en términos de facturación y posicionamiento comercial, lo que las convierte en unidades más estratégicas para mantener dentro de la cadena.

Por lo tanto, la Tienda 4 representa la opción más coherente para ser vendida, permitiendo al Sr. Juan liberar recursos para invertir en su nuevo emprendimiento sin comprometer el rendimiento global de Alura Store.
```
